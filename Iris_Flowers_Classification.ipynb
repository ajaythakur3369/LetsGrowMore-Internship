## **Iris Flower Classification Using Decision Tree Algorithm:**

The goal of the project is to classify different species of iris flowers based on their petal and sepal dimensions. The dataset contains 150 samples of iris flowers, each consisting of four features: sepal length, sepal width, petal length, and petal width. The target variable is the species of iris flower, which can take three possible values: setosa, versicolor, and virginica. The project involves building a machine learning model that can accurately classify new samples of iris flowers based on their features, using various classification algorithms such as logistic regression, decision trees, k-nearest neighbors, support vector machines, or neural networks. In this project, we are classifying using the Decision Tree Algorithm.

The Iris Flowers dataset contains 150 samples of iris flowers, where each sample contains four features and a target variable indicating the species of the flower. The details of the variables in the dataset are as follows:

* Sepal Length (in cm): Represents the length of the sepal of the iris flower.
* Sepal Width (in cm): Represents the width of the sepal of the iris flower.
* Petal Length (in cm): Represents the length of the petal of the iris flower.
* Petal Width (in cm): Represents the width of the petal of the iris flower.
* Species: Represents the species of the iris flower, which can be one of three possible values: setosa, versicolor, or virginica.
* The target variable, "Species," is a categorical variable represented as a string.

## **Splitting the dataset:**

```python
x = df.drop('Species', axis = 1)
y = df['Species']
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest= train_test_split(x, y, test_size = 0.3, random_state = 42)
```

## **Importing Decision Tree Classifier:**

```python
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier()
```

```python
dtc.fit(xtrain,ytrain)
ypred = dtc.predict(xtest)
```

## **Model Evaluation:**

```python
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
```

```python
print(f'Accuracy Score: {accuracy_score(ytest, ypred)}')
```

```python
print("Classification_Report\n", classification_report(ypred, ytest))
```

## **Checking for Overfitting and Underfitting:**

```python
train = dtc.score(xtrain, ytrain)
test = dtc.score(xtest, ytest)
print(f"Training Accuracy: {train}\nTesting Accuracy: {test}")
```

**Perfect model**

## **Scaling:**

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
xs_train = scaler.fit_transform(xtrain)
xs_test = scaler.transform(xtest)
```

```python
dtcs = DecisionTreeClassifier()
dtcs.fit(xs_train, ytrain)
ypred = dtcs.predict(xs_test)
print(f'Accuracy Score: {accuracy_score(ytest, ypred)}')
print("Classification_report\n", classification_report(ypred, ytest))
```

```python
train = dtcs.score(xs_train, ytrain)
test = dtcs.score(xs_test, ytest)
print(f"Training Accuracy: {train}\n Testing Accuracy: {test}")
```

## **Hyperparameter Tunning:**

```python
from sklearn.model_selection import GridSearchCV
```

```python
dtc = DecisionTreeClassifier()
param_grid = {
    "criterion":["gini", "entropy"],
    "max_depth":list(range(1, 20)),
    "min_samples_leaf":list(range(1, 20))
}
grid = GridSearchCV(estimator = dtc, param_grid = param_grid, verbose = 2)
grid.fit(xs_train, ytrain)
ypred = grid.predict(xs_test)
```

```python
grid.best_params_
```

```python
print(f'Accuracy Score: {accuracy_score(ytest, ypred)}')
train = grid.score(xs_train, ytrain)
test = grid.score(xs_test, ytest)
print(f'Training Accuracy: {train}\n Testing Accuracy: {test}')
print()
print("Classification_report\n",classification_report(ypred, ytest))
```
