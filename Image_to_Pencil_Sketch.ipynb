## **Displaying a Real Image:**

```python
img = cv.imread('/content/drive/MyDrive/Colab_Notebook/Internship_Name/LetsGrowMore/File_Name/Lion_image.jpg', 1)
img = cv.resize(img, (640, 400))
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
display(Image.fromarray(img))
```

## **Grayscale Image:**

```python
gray_filter = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
```

```python
cv2_imshow(gray_filter)
```

## **Negative Image:**

```python
neg_img = cv.bitwise_not(gray_filter)
display(Image.fromarray(neg_img))
```

## **Blurred Image:**

```python
blur_img = cv.GaussianBlur(neg_img,(21, 21), 0)
display(Image.fromarray(blur_img))
```

## **Inverted Blurred Image:**

```python
inv_blur = cv.bitwise_not(blur_img)
display(Image.fromarray(inv_blur))
```

## **Finally, Pencil Sketch Image:**

```python
sketch = cv.divide(gray_filter, inv_blur, scale = 190)
display(Image.fromarray(sketch))
```
